```python
import model_one
import utils_one
import config
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
import shap

Target_Indices = [273, 275, 277, 281, 283, 285]
target_col = ['C21C05_01B2', 'C21C05_01C2', 'C21C05_01D2', 'C21C05_01F2', 'C21C05_01G2', 'C21C05_01H2']
#target_col = ['C21C05_01B2']
FI_lst = []
num = -1
colors = ['darkred', 'green', 'blue', 'orange', 'darkgrey', 'black']


for idx, val in enumerate(target_col):
    
    num+=1
    
    df, meta = utils_one.data_import()
    print(f"\n==============================================\n CLASS: {meta.column_labels[Target_Indices[idx]]}\n==============================================\n\n")
    X, Y = utils_one.data_processing(val, num)
    
    X_resampled, y_resampled = utils_one.data_aug_smote(X, Y)
    XGB_model, X_test = model_one.XGBoost_for_all(X_resampled, y_resampled)
    FI_vals = model_one.feature_importance(XGB_model, X, df, meta)
    FI_lst.append(FI_vals)
    
    explainer = shap.TreeExplainer(XGB_model)
    shap_values = explainer.shap_values(X_test)
    
    sv = explainer(X_test)

    mean_values = np.mean(sv.values, axis=0)#.reshape(1, 22)
    feature_names = sv.feature_names
    feature_names = [s.replace('C21', '') for s in feature_names]

    bars = plt.bar(np.arange(len(mean_values)), mean_values, color=colors[num])
    plt.xticks(np.arange(len(mean_values)), feature_names, rotation=45, ha='right')
    
    #for bar in bars:
    #    yval = bar.get_height()  # Get the height of the bar
    #    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')
    
    plt.xlabel('Features', fontsize=13)
    plt.ylabel('Mean Values', fontsize=13)
    plt.title(f'{val}', fontsize=16)
    plt.grid(True, linestyle='--', color='lightgray', linewidth=0.7)
    plt.show()
```
