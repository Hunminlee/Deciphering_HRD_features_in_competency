```python
import model_one
import utils_one
import config
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def main():
    FI_lst = []
    
    for idx, val in enumerate(config.target_col):    
        df, meta = utils_one.data_import()
        print(f"\n==============================================\n CLASS: {meta.column_labels[config.Target_Indices[idx]]}\n==============================================\n\n")
        X, Y = utils_one.data_processing(val, idx)
        #X = HRD_col_select(all_X)
        X_resampled, y_resampled = utils_one.data_aug_smote(X, Y)
        XGB_model = model_one.XGBoost_for_all(X_resampled, y_resampled)
        print("\n")
        FI_vals = model_one.feature_importance(XGB_model, X, df, meta)
        FI_lst.append(FI_vals)
        
    return FI_lst

def sort_df(FI_LST):
    combined_df = pd.concat(FI_LST)
    result_df = combined_df.groupby('Feature', as_index=False).sum()
    sorted_result_df = result_df.sort_values(by='Importance', ascending=False)
    print(f'\n\nSorted feature and their importance\n==================================================\n{sorted_result_df}')
```

```python
if __name__ == "__main__":
    FI_LST = main()
    sort_df(FI_LST)
```
